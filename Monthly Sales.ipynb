```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_excel('Sales_data_pythonZ.xlsx')
```

```python
Sales_by_Product=df.groupby('product_name')['total_sales'].sum().reset_index()
```

```python
Sales_by_Product
```

```python
product_sales_sorted = Sales_by_Product.sort_values(by='total_sales', ascending=False)
product_sales_sorted

```

```python
channel_profit = df.groupby('channel')['profit'].sum()
plt.figure(figsize=(8, 8))
plt.pie(channel_profit, labels=channel_profit.index, autopct='%1.1f%%', startangle=140)
plt.title('Profit Percentage by Channel')
plt.tight_layout()
plt.show()

```

```python

product_sales = df.groupby('product_name')['total_sales'].sum().reset_index()
top_products = product_sales.sort_values(by='total_sales', ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(data=top_products, x='total_sales', y='product_name', palette='viridis')
plt.title('Products by Total Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['profit_margin_pct'], bins=20, kde=True, color='blue')
plt.title('Profit Margin % Distribution')
plt.xlabel('Profit Margin (%)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
us_region_sales = df.groupby('us_region')['total_sales'].sum()
plt.figure(figsize=(7, 5))
sns.barplot(x=us_region_sales.index, y=us_region_sales.values, palette='deep')
plt.title('Total Sales by US Region')
plt.xlabel('US Region')
plt.ylabel('Total Sales')
plt.tight_layout()

plt.show()
```

```python
weekly_sales = df.groupby('weeknum')['total_sales'].sum().sort_index()
plt.figure(figsize=(8, 5))
weekly_sales.plot(marker='o', kind='line', color='navy')
plt.title('Weekly  Sales Trend')
plt.xlabel('Week Number')
plt.ylabel('Total Sales')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='channel', y='total_sales', estimator=sum, ci=None, palette='deep')

plt.title('Total Sales by Channel')
plt.ylabel('Total Sales')
plt.xlabel('Channel')
plt.tight_layout()
plt.show()
```
